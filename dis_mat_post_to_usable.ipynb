```python
with open("dis_mat_post.json", "r") as file:
    contents = file.read()
    print(contents)
```

```python
import json
data = json.loads(contents)
```

```python
data
```

```python
data = [json.loads(datum) for datum in data]
```

```python
data[0]['results'][0]
```

```python
result = {}
for responds in data:
    for onexn_mat in responds['results']:
        result[int(onexn_mat['search_id'].split('.')[1])] = {}
        for location in onexn_mat['locations']:
            if int(onexn_mat['search_id'].split('.')[1]) != 75 and int(location['id'].split('.')[1]) != 75:
                result[int(onexn_mat['search_id'].split('.')[1])][int(location['id'].split('.')[1])] = location['properties'][0]['travel_time'] / 60
```
